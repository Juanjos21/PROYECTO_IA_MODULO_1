## Descripción del dataset

El dataset **salud_mundial.csv** contiene información sobre indicadores de salud de aproximadamente 159 países a nivel mundial.

Incluye variables como:

* Esperanza de vida
* Gasto en salud por persona
* Mortalidad infantil
* Número de médicos por cada 1000 habitantes
* Porcentaje de vacunación
* Nivel de ingresos de cada país
* Otros indicadores relacionados con salud pública

Este tipo de dataset permite analizar las condiciones de salud global y comparar regiones del mundo.

## Cómo ejecutar el proyecto

Sigue estos pasos para ejecutar el notebook correctamente:

1. Descargar o clonar el repositorio
2. Abrir el proyecto en Visual Studio Code o Jupyter Notebook
3. Asegurarse de que el archivo `salud_mundial.csv` esté en la misma carpeta del notebook
4. Instalar las librerías necesarias:

   ```bash
   pip install pandas
   ```
5. Abrir el archivo `.ipynb`
6. Ejecutar las celdas en orden (Run All o ejecutar una por una)

## Hallazgos y conclusiones

1. Existe una relación positiva entre el gasto en salud y la esperanza de vida en muchos países.
2. Los países con niveles de ingresos altos tienden a tener menor mortalidad infantil.
3. Algunas regiones presentan mayores niveles de obesidad, lo que puede estar relacionado con el estilo de vida y desarrollo económico.
